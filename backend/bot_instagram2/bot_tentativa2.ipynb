```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from PIL import Image, ImageDraw, ImageFont
from instabot import Bot as InstaBot
import os
import time
import random

# Configuração do ChromeDriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')  # Executa o navegador em modo headless (sem interface gráfica)
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Substitua 'C:/Users/edeli/OneDrive/Área de Trabalho/chromedriver-win64/chromedriver.exe' pelo caminho real onde o chromedriver.exe está localizado
service = Service("chromedriver.exe")
browser = webdriver.Chrome(service=service)

# Função para adicionar texto na imagem
def adicionar_texto_na_imagem(caminho_imagem, texto, caminho_saida):
    imagem = Image.open(caminho_imagem)
    draw = ImageDraw.Draw(imagem)
    fonte = ImageFont.load_default()
    posicao_texto = (10, 10)
    draw.text(posicao_texto, texto, font=fonte, fill="white")
    imagem.save(caminho_saida)

# Função para escolher uma imagem aleatória da pasta
def escolher_imagem_aleatoria(pasta):
    imagens = [os.path.join(pasta, f) for f in os.listdir(pasta) if os.path.isfile(os.path.join(pasta, f))]
    return random.choice(imagens)

# Função para realizar login no LinkedIn e buscar vagas afirmativas para mulheres
def buscar_vagas(browser):
    browser.get("https://www.linkedin.com/login")
    time.sleep(2)

    email_input = browser.find_element(By.NAME, "session_key")
    email_input.send_keys('marina.sfigueiredo39@gmail.com')
    password_input = browser.find_element(By.NAME, "session_password")
    password_input.send_keys("Msf20032003")
    password_input.send_keys(Keys.RETURN)
    time.sleep(5)

    browser.get("https://www.linkedin.com/jobs/search?keywords=vagas%2Bpara%2Bmulheres%2Bna%2B%C3%A1rea%2Bde%2Btecnologia&location=Brasil&geoId=106057199&trk=public_jobs_jobs-search-bar_search-submit&currentJobId=3863620932&position=2&pageNum=0")
    time.sleep(5)  # Aguarda o carregamento da página

    # Adiciona uma espera explícita para garantir que o elemento esteja presente
    wait = WebDriverWait(browser, 10)
    qtd_vagas = wait.until(EC.presence_of_element_located((By.CLASS_NAME, "results-context-header__job-count")))
    print("Quantidade de vagas: " + qtd_vagas.text)

    # Transformar para inteiro
    qtd_vagas_int = int(qtd_vagas.text.replace('.', ''))
    print(f"Quantidade de vagas: {qtd_vagas_int}")

# Pasta de saída para as imagens
pasta_saida = "/Users/edeli/OneDrive/Área de Trabalho/imagens_saida"

# Certifique-se de que a pasta de saída existe
if not os.path.exists(pasta_saida):
    os.makedirs(pasta_saida)

# Inicializar a lista de vagas
lista = [("Desenvolvedora Backend", "Empresa X", "São Paulo"),
         ("Engenheira de Software", "Empresa Y", "Rio de Janeiro"),
         ("Analista de Dados", "Empresa Z", "Belo Horizonte")]

# Pasta das imagens
pasta_imagens = "/Users/edeli/OneDrive/Área de Trabalho/imagens_bot"

# Buscar vagas no LinkedIn
buscar_vagas(browser)
browser.quit()

class BotInstagram:

    def __init__(self):
        service = Service("chromedriver.exe")
        self.driver = webdriver.Chrome(service=service)
        self.instabot = InstaBot()

    def entrar_link(self, link):
        self.driver.get(link)

    def login(self, username, password):
        try:
            # Esperar até que os campos de login estejam presentes
            WebDriverWait(self.driver, 10).until(
                EC.presence_of_element_located((By.NAME, "username"))
            )
            username_input = self.driver.find_element(By.NAME, "username")
            password_input = self.driver.find_element(By.NAME, "password")

            # Inserir o nome de usuário e senha
            username_input.send_keys(username)
            password_input.send_keys(password)

            # Submeter o formulário
            password_input.send_keys(Keys.RETURN)

            time.sleep(10)

            print("Login bem-sucedido!")
        except Exception as e:
            print(f"Erro ao tentar fazer login: {e}")

    def upload_foto(self, pasta_imagens, texto):
        try:
            # Escolher uma imagem aleatória
            #caminho_imagem = escolher_imagem_aleatoria(pasta_imagens)
            caminho_saida = os.path.join(pasta_imagens, "imagem_post.jpg")
            caminho_da_imagem = "/Users/edeli/OneDrive/Área de Trabalho/imagens_bot/grl_pwr.jpg"
            imagem = Image.open(caminho_da_imagem)
            
            # Adicionar texto na imagem
            adicionar_texto_na_imagem(caminho_da_imagem, texto, caminho_saida)
            
            # Fazer upload da imagem usando instabot
            self.instabot.upload_photo(caminho_saida, caption=texto)
            print("Foto carregada com sucesso!")
        except Exception as e:
            print(f"Erro ao carregar a foto: {e}")

    def logout(self):
        try:
            self.instabot.logout()
            print("Logout bem-sucedido!")
        except Exception as e:
            print(f"Erro ao fazer logout: {e}")



bot = BotInstagram()
bot.entrar_link("https://www.instagram.com/")
bot.login(username='bot_insta2025', password='msf2025')
time.sleep(10)
# bot.upload_photo(escolher_imagem_aleatoria(pasta_imagens))
texto_para_post = "Confira as últimas vagas afirmativas para mulheres na tecnologia!"
bot.upload_foto(pasta_imagens, texto_para_post)
# bot.fazer_posts(pasta_imagens, pasta_saida, lista)

bot.logout()
```
